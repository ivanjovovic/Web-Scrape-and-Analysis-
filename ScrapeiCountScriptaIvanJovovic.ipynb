Scrapovanje, ciscenje teksta i cuvanje podataka u txt fajl

```python
from bs4 import BeautifulSoup
import requests
import re
urls= ['https://fist.udg.edu.me/predavaci/','https://fmefb.udg.edu.me/predavaci/','https://fkt.udg.edu.me/predavaci/','https://politehnika.udg.edu.me/predavaci/','https://fpn.udg.edu.me/predavaci/','https://hs.udg.edu.me/predavaci/','https://fu.udg.edu.me/predavaci/','https://fptbhe.udg.edu.me/predavaci/','https://fsm.udg.edu.me/predavaci/','https://fdm.udg.edu.me/predavaci/','https://ff.udg.edu.me/predavaci/','https://fprn.udg.edu.me/predavaci/']
for url in urls:
  text=requests.get(url).text
  soup=BeautifulSoup(text,"html5lib")
  predavaci=soup.find_all("a",{"class":"teacher-name"})
  predavac = str(predavaci)
  cisto=re.sub('<[^>]+>',"", predavac)
  cisto1 = cisto.replace('[', "")
  cisto2 = cisto1.replace(']', "")
  cisto3 = cisto2.replace(',', "")
  cisto3=re.sub('ć',"c", cisto3)
  cisto3=re.sub('Ć',"C", cisto3)
  cisto3=re.sub('č',"c", cisto3)
  cisto3=re.sub('Č',"C", cisto3)
  cisto3=re.sub('š',"s", cisto3)
  cisto3=re.sub('Š',"S", cisto3)
  cisto3=re.sub('ž',"z", cisto3)
  cisto3=re.sub('Ž',"Z", cisto3)
  print(cisto3)
  text_file = open("profesori1.txt", "a")
  n = text_file.write(cisto3)
  text_file.close()

```

Uklanjanje duplikata i white spaces iz prethodno generisanog txt fajla

```python
duplikati = set() 

with open("profesori1.txt", "r+") as f:
    d = f.readlines()
    f.seek(0)
    for i in d:
        if i not in duplikati:
            f.write(i)
            duplikati.add(i)
    f.truncate()
```

Racunanje broja predavaca

```python
with open("profesori2.txt") as lines:
  profesori=0
  docenti=0
  doktori=0
  magistri=0
  ostali=0
  for line in lines:
    if line.startswith("Prof."):
      profesori+=1
    elif line.startswith("Doc."):
      docenti+=1
    elif line.startswith("dr"):
      doktori+=1
    elif line.startswith("mr"):
      magistri+=1
    else:
      ostali+=1

  print("Broj profesora je:" + str(profesori))
  print("Broj docenta je:" + str(docenti))
  print("Broj doktora je:" + str(doktori))
  print("Broj magistara je:" + str(magistri))
  print("Broj ostalih je:" + str(ostali))

```
